```python
## Fazendo as duas tarefas teríamos esse código
dados_tarefa = {
    "nome": ["Ana", "Beatriz", "Carla", "Diego"],
    "idade": [16, 17, 16, 20],
    "salario": [1000, 2000, 1515, 5000]
}

df_tarefa = pd.DataFrame(dados_tarefa)

sns.barplot(x="idade", y="salario", data=df_tarefa)
plt.show()
```

```python
## Introduzindo if e else

## Variáveis booleanas (True ou False)
maior_de_idade = df_tarefa["idade"][0] >= 18        # Ao usar o índice 0 estamos selecionando os dados da Ana que é o primeiro dado na variável dados_tarefa
bom_salario = df_tarefa["salario"][0] >= 3000       # Podemos trocar o 0 por 1, 2 ou 3 já que o conjunto de dados tem 4 dados e Python conta a partir do 0

## Variáveis booleanas, lógica combinada
adulto_independente = maior_de_idade and bom_salario             # Se maior_de_idade = True e bom_salario = True, então adulto_independente = True
jovem_dependente = (not maior_de_idade) and (not bom_salario)    # Se maior_de_idade = False e bom_salario = False, então jovem_dependente = True

if adulto_independente:
    print("Adulto bem de vida")
elif jovem_dependente:
    print("Jovem sem dinheiro")
else:
    print("Nenhuma das opções")
```
