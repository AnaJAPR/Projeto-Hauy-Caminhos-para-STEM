```python
# Exemplo base antes da tarefa
nomes = ["Alice", "João", "Ana", "José", "Carlos"]
nomes_com_a = []
for nome in nomes:
    if nome.startswith("A"):
        nomes_com_a.append(nome)

print(nomes_com_a)
```

```python
''' TAREFA 1 feita em sala
Adicione Janaína, Caio e Cristiane na lista de nomes.
Crie 2 listas vazias para nomes que começam com C e J.
Use for para adicionar todos os nomes que começam com C e J nessas listas.
Imprima as listas no terminal.
'''

# Resposta
lista_nomes = ["Alice", "João", "Ana", "José", "Carlos", "Janaína", "Caio", "Cristiane"]
nomes_com_c = []
nomes_com_j = []

for nome in lista_nomes:
    if nome.startswith("C"):
        nomes_com_c.append(nome)
    elif nome.startswith("J"):
        nomes_com_j.append(nome)

print(nomes_com_c)
print(nomes_com_j)
```

```python
'''TAREFA 2 feita em sala
Crie uma lista "idades" com os elementos 17, 18, 25, 43, 15, 29, 36 e 50.
Usando for e if crie outra lista com todas as idades maiores ou iguais a 18.
Crie uma terceira lista com as idades menores que 18 também.
'''

# Resposta

idades = [17, 18, 25, 43, 15, 29, 36, 50]
maiores_de_idade = []
menores_de_idade = []

for idade in idades:
    if idade >= 18:
        maiores_de_idade.append(idade)
    else:
        menores_de_idade.append(idade)

print(maiores_de_idade)
print(menores_de_idade)
```

```python
# Exemplo base antes da tarefa
salarios = [2000, 5000, 1500, 2100, 5000, 1500, 2000]
contagem = {}    # dicionário vazio

for salario in salarios:
    if salario in contagem:
        contagem[salario] += 1
    else:
        contagem[salario] = 1

print(contagem)
```

```python
'''TAREFA 3 feita em sala
Crie uma lista com os valores A, B, C, C, A, C, B, B, A.
Defina um dicionário vazio chamado contagem_categorias.
Usando for e if/else, conte quantas vezes cada categoria aparece na lista e adicione isso no dicionário contagem_categorias.
Imprima o dicionário no terminal.
'''

# Resposta
categorias = ["A", "B", "C", "C", "A", "C", "B", "B", "A"]
contagem_categorias = {}

for categoria in categorias:
    if categoria in contagem_categorias:
        contagem_categorias[categoria] += 1
    else:
        contagem_categorias[categoria] = 1

print(contagem_categorias)
```

```python
# Exemplo base antes da tarefa
numero = 1
while numero <= 10:
    if numero % 5 == 0:
        print(f"Número {numero} é divisível por 5")
    numero += 1
```

```python
'''TAREFA 4 feita em sala
Faça um while como o anterior, mas a condição de parada agora deve ser número menor que 20.
Imprima "Número par" quando for par e "Número ímpar" quando for ímpar.
'''

# Resposta
numero = 1
while numero < 20:
    if numero % 2 == 0:
        print(f"O número {numero} é par.")
    else:
        print(f"O número {numero} é ímpar.")
    numero += 1
```
